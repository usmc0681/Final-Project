```python
import pickle
import joblib
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
```

```python
sc = MinMaxScaler()
mdl = joblib.load('recidivism_model.pkl')
X = pd.read_csv('small_rec_dataset.csv')
X_test = sc.fit_transform(X)
y_pred = mdl.predict(X_test)
y_pred = (y_pred > 0.5).astype(np.float32)
```

```python
val = pd.DataFrame(y_pred, columns = ['Redicivism Risk'])
val = val.replace({1: 'High Risk', 0: 'Low Risk'})
val
```
